```python
import json
with open('movies(5).json', 'r') as f:
    db = f.read()
    db = json.loads(db)
```

```python
db
```

```python
ext_data = 0
need_data = 0

for movie in db['movies']:
    if 'plot' in movie.keys():
        ext_data += 1
    else:
        need_data += 1

total_movies = len(db['movies'])
percentage_need_data = (need_data / total_movies) * 100

print(f"Percentage of movies that need data: {percentage_need_data:.2f}%")
```
